```python

def generate_random_feature_value(center, stddev=0.05, feature_type='continuous'):
    if feature_type == 'continuous':
        return np.random.normal(loc=center, scale=stddev)
    elif feature_type == 'discrete':
        return int(np.round(np.random.normal(loc=center, scale=stddev)))
    else:
        raise ValueError("Invalid feature_type. Choose 'continuous' or 'discrete'.")

def calculate_likelihood(value, center, stddev=0.05, feature_type='continuous'):
    """
    Calculate the likelihood of a given value in a specified normal distribution.

    Parameters:
    - value: The value for which to calculate the likelihood.
    - center: The mean (center) of the distribution.
    - stddev: The standard deviation of the distribution. Default is 0.05.
    - feature_type: The type of feature ('continuous' or 'discrete').

    Returns:
    - The likelihood (density) of the given value in the distribution.
    """
    if feature_type == 'continuous':
        # For continuous features, directly use the PDF of the normal distribution.
        likelihood = norm.pdf(value, loc=center, scale=stddev)
    elif feature_type == 'discrete':
        # For discrete features, calculate the PDF at the nearest integer.
        discrete_value = int(np.round(value))
        likelihood = norm.pdf(discrete_value, loc=center, scale=stddev)
    else:
        raise ValueError("Invalid feature_type. Choose 'continuous' or 'discrete'.")
    
    return likelihood

# function to generate a random value for each feature
def generate_random_feature_values(centers, std_devs):
    random_values = {}
    for feature, center in centers.items():
        feature_type = 'discrete' if feature in ['key', 'mode'] else 'continuous'
        random_values[feature] = generate_random_feature_value(center, stddev=std_devs[feature], feature_type=feature_type)
    return random_values


def calculate_all_log_likelihoods(centers, std_devs, values):
    """
    Assuming that all features are independent, calculate the likelihood of the current value. 
    """
    total_likelihood = 1
    for feature, center in centers.items():
        feature_type = 'discrete' if feature in ['key', 'mode'] else 'continuous'
        curr_likelihood = calculate_likelihood(value=values[feature], center=center, stddev=std_devs[feature], feature_type=feature_type)

        total_likelihood += (np.log(curr_likelihood))
        # print(f"curr_likelihood: {np.log(curr_likelihood)}, feature: {feature}, center, val: {center},{values[feature]}")
    return total_likelihood


def calculate_log_likelihood_for_row(row, true_centers, std_devs):
    values = {feature: row[feature] for feature in true_centers.keys()}
    return calculate_all_log_likelihoods(true_centers, std_devs, values)


def compute_stats(df, parameters):
    """
    Compute the mean and standard deviation for specified parameters in a DataFrame.

    Parameters:
    - df: Pandas DataFrame containing the data.
    - parameters: List of strings, names of the columns for which to compute statistics.

    Returns:
    - A dictionary with each parameter as keys and their mean and standard deviation as values.
    """
    stats = {}
    for param in parameters:
        if param in df.columns:  # Check if the parameter is in the DataFrame
            mean = df[param].mean()
            std_dev = df[param].std()
            stats[param] = {'mean': mean, 'std_dev': std_dev}
        else:
            print(f"Warning: '{param}' not found in DataFrame columns.")

    return stats



def compute_cluster_weighted_parameters(num_songs_each_cluster, parameters):
    # Compute the weight for each cluster
    cluster_weights = [num_songs / sum(num_songs_each_cluster) for num_songs in num_songs_each_cluster]
    
    # Initialize a dictionary to store the computed means for each parameter
    parameter_means = {}
    
    for parameter in parameters:
        # weighted_sum = 0
        # total_weight = 0  # To account for clusters with 0 tracks
        resampled_vals = []
        
        for cluster, weight in enumerate(cluster_weights):
            # Extract parameter values for the current cluster
            cluster_param_values = song_df[song_df['cluster'] == cluster][parameter]
            
            # Sample from this cluster's parameter values based on the weight
            # Here we assume the number of samples to draw proportional to the total weight
            samples = np.random.choice(cluster_param_values, size=int(weight * 10000), replace=True)
        
        resampled_vals.append(samples)
        parameter_means[parameter] = np.mean(resampled_vals)   
                # weighted_sum += np.mean(samples) * weight
                # total_weight += weight  
        # Compute the weighted mean for the parameter
        # if total_weight > 0:
        #     parameter_means[parameter] = weighted_sum / total_weight
        # else:
        #     parameter_means[parameter] = None  # No data for this parameter

    keys_to_drop = ['key', 'time_signature', 'mode', 'popularity']
# keys = set(parameter_means.keys()).difference(set(keys_to_drop))


    modified_input_parameters = {key: value for key, value in parameter_means.items() if key not in keys_to_drop}

    # for key in keys:
    #     if key == 'tempo':
    #         continue 
        
    #     modified_input_parameters.update({'min_' + key : parameter_means[key] - 0.1})
    #     modified_input_parameters.update({'max_' + key : parameter_means[key] + 0.1})
    
    return modified_input_parameters

def update_num_songs_each_cluster(old_vector, track_ids):
    features_col = ['danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', "time_signature"]
    
    if old_vector == [0 for _ in range(20)]:
        plus_three = True
    else:
        plus_three = False
        
    def classify_new_songs(track_ids=track_ids, features_col=features_col):
        features_list = sp.audio_features(tracks=track_ids)
        features_df = pd.DataFrame(features_list, columns=features_col)
        processed_features_df = rs.transform(features_df)
        
        clustering_labels = kmeans.predict(processed_features_df)
        
        return clustering_labels

    labels = classify_new_songs(track_ids, features_col)
    for cluster in labels:
        old_vector[cluster] += 1 if plus_three == False else 3
    
    return old_vector     

# List of parameters to compute weighted means for
parameters = ['danceability', 'energy',
              'loudness', 'speechiness', 'acousticness',
              'instrumentalness', 'liveness', 'valence', 'tempo']
    
# Example usage
tracks_in_each_cluster = [0 for _ in range(19)]
track_ids = ['5SuOikwiRyPMVoIQDJUgSV', '4qPNDBW1i3p13qLCt0Ki3A', '1iJBSr7s7jYXzM8EGcbK5b']  # Add more track_ids as needed

tracks_in_each_cluster = update_num_songs_each_cluster(tracks_in_each_cluster, track_ids)
weighted_parameter_means = compute_cluster_weighted_parameters(tracks_in_each_cluster, parameters)
print(weighted_parameter_means)
```
